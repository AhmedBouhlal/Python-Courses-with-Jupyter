## ðŸ§  Concept: What is Matplotlib?
Matplotlib is the most powerful Python library for 2D plotting. Itâ€™s used to create everything from simple line plots to complex dashboards.
    The core module for plotting is:

```python
import matplotlib.pyplot as plt
```

### âœ… PART 1 â€” Your First Plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.plot(x, y)       # Plot line
plt.show()           # Show the plot
```

### âœ… PART 2 â€” Add Labels, Title, and Grid

```python
plt.plot(x, y)
plt.title("My First Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.show()
```

ðŸ§  .title(), .xlabel(), and .ylabel() help make your plot understandable.
### âœ… PART 3 â€” Customize Line Style, Color, and Marker

```python
plt.plot(x, y, color="red", linestyle="--", marker="o", label="Line A")
plt.legend()
plt.title("Styled Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

| Option      | Example         | Meaning                |
| ----------- | --------------- | ---------------------- |
| `color`     | `"blue"`        | Line color             |
| `linestyle` | `"--"` or `":"` | Dashed / dotted        |
| `marker`    | `"o"`           | Circle markers         |
| `label`     | `"Line A"`      | Used in legend display |


### âœ… PART 4 â€” Save Your Plot to a File

```python
plt.plot(x, y)
plt.title("Saved Plot")
plt.savefig("my_plot.png")  # Save as image
plt.show()
```

ðŸ”¥ You can also save as:

```python
plt.savefig("plot.svg")  # Vector format
plt.savefig("plot.pdf")  # PDF for print
```

### âœ… PART 5 â€” Multiple Lines in One Plot

```python
x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [2, 3, 6, 12]

plt.plot(x, y1, label="y = x^2", color="green")
plt.plot(x, y2, label="y = 3x", color="orange")
plt.title("Multiple Lines")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

### âœ… PART 6 â€” Add Axis Limits and Ticks

```python
plt.plot(x, y1)
plt.title("Axis Control")
plt.xlim(0, 5)
plt.ylim(0, 20)
plt.xticks([1, 2, 3, 4])
plt.yticks([0, 5, 10, 15, 20])
plt.grid(True)
plt.show()
```

## ðŸ”¶ STAGE 2 â€“ Charts & Layouts in Matplotlib
### âœ… PART 1 â€” Bar Charts (plt.bar())
Bar charts are great for comparing values between categories.

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [10, 25, 8, 15]

plt.bar(categories, values, color='skyblue')
plt.title("Bar Chart Example")
plt.xlabel("Category")
plt.ylabel("Value")
plt.grid(axis='y')
plt.show()
```

ðŸ“Œ axis='y' enables horizontal grid lines.
### âœ… PART 2 â€” Horizontal Bar Charts (plt.barh())

```python
plt.barh(categories, values, color='lightgreen')
plt.title("Horizontal Bar Chart")
plt.xlabel("Value")
plt.ylabel("Category")
plt.grid(axis='x')
plt.show()
```

### âœ… PART 3 â€” Histogram (plt.hist())
Histograms are for distributions (e.g. grades, heights, AI confidence scores).

```python
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=30, color='purple', alpha=0.7)
plt.title("Histogram of Random Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

| Param            | Meaning        |
| ---------------- | -------------- |
| `bins=30`        | Number of bars |
| `alpha=0.7`      | Transparency   |
| `color="purple"` | Color of bars  |

### âœ… PART 4 â€” Subplots with plt.subplot()
This allows multiple plots in one figure.

```python
x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [1, 2, 3, 4]

plt.subplot(1, 2, 1)
plt.plot(x, y1, color='red')
plt.title("Plot 1")

plt.subplot(1, 2, 2)
plt.plot(x, y2, color='blue')
plt.title("Plot 2")

plt.tight_layout()
plt.show()
```

ðŸ“Œ subplot(1, 2, 1) = 1 row, 2 columns, plot #1

### âœ… PART 5 â€” Subplots with plt.subplots()
This is the modern & flexible way to build grid layouts:

```python
fig, axs = plt.subplots(2, 2, figsize=(6, 6))

axs[0, 0].bar(['A', 'B'], [5, 7])
axs[0, 0].set_title("Bar")

axs[0, 1].plot([1, 2, 3], [3, 2, 5])
axs[0, 1].set_title("Line")

axs[1, 0].scatter([1, 2, 3], [3, 1, 4])
axs[1, 0].set_title("Scatter")

axs[1, 1].hist(np.random.randn(100), bins=20)
axs[1, 1].set_title("Histogram")

fig.suptitle("2x2 Subplot Example")
plt.tight_layout()
plt.show()
```

## ðŸŽ¨ STAGE 3 â€“ Styling, Themes, and Global Settings
### âœ… PART 1 â€” Change Font Size, Font Type, and Title Styles
ðŸ”§ Customize title, labels, ticks:

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.plot(x, y, color='green')
plt.title("Styled Title", fontsize=18, fontweight='bold', color='purple')
plt.xlabel("X Axis", fontsize=12)
plt.ylabel("Y Axis", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(True)
plt.show()
```

### âœ… PART 2 â€” Use Built-in Styles / Themes
Matplotlib has built-in themes like "ggplot", "seaborn", "dark_background".

```python
plt.style.use("ggplot")

plt.plot(x, y)
plt.title("Using ggplot Style")
plt.show()
```

### âœ… PART 3 â€” Global Styling with rcParams
plt.rcParams changes default behavior for all future plots.

```python
plt.rcParams["figure.figsize"] = (6, 4)
plt.rcParams["lines.linewidth"] = 2
plt.rcParams["axes.titlesize"] = 16
plt.rcParams["axes.labelsize"] = 12
plt.rcParams["axes.facecolor"] = "#f4f4f4"
plt.rcParams["font.family"] = "monospace"
```

Then:

```python
plt.plot(x, y)
plt.title("Global Style Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

### âœ… PART 4 â€” Dark Mode Plotting ðŸŒ’

```python
plt.style.use("dark_background")

x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y, color="cyan", marker="o")
plt.title("Dark Mode Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

ðŸ§  Works great for presentations or dashboards.
### âœ… PART 5 â€” Custom Color Maps (for heatmaps, scatter, etc.)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)

plt.scatter(x, y, c=colors, cmap="viridis", s=100, alpha=0.7)
plt.title("Custom Colormap with Colorbar")
plt.grid(False)           # ðŸ‘ˆ Disable grid before colorbar
plt.colorbar()            # Now it's clean
plt.show()
```

## ðŸ”· STAGE 4 â€“ Advanced Plot Types
### âœ… PART 1 â€” Show Images with imshow()
Use imshow() to display 2D data as an image (heatmap style)

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.rand(10, 10)

plt.imshow(data, cmap="viridis", interpolation="nearest")
plt.title("Heatmap from Matrix")
plt.grid(False)
plt.colorbar()
plt.show()
```

| Param           | Description                                             |
| --------------- | ------------------------------------------------------- |
| `cmap`          | Color scheme (`"viridis"`, `"hot"`, `"coolwarm"`, etc.) |
| `interpolation` | Pixel smoothing method (`"nearest"` for raw)            |

### âœ… PART 2 â€” Real Heatmap with Labels

```python
import seaborn as sns
import pandas as pd

data = np.random.rand(6, 6)
df = pd.DataFrame(data, columns=list('ABCDEF'))

sns.heatmap(df, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Labeled Heatmap")
plt.show()
```

### âœ… PART 3 â€” Contour & Filled Contour Plots

```python
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X**2 + Y**2)

# Contour lines
plt.contour(X, Y, Z, levels=20, cmap="plasma")
plt.title("Contour Plot")
plt.colorbar()
plt.show()

# Filled contour
plt.contourf(X, Y, Z, levels=20, cmap="plasma")
plt.title("Filled Contour")
plt.colorbar()
plt.show()
```

### âœ… PART 4 â€” 3D Plots with Axes3D

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap="viridis")
ax.set_title("3D Surface Plot")
plt.show()
```

### âœ… BONUS â€” 3D Scatter Plot

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

ax.scatter(x, y, z, c=z, cmap='plasma')
ax.set_title("3D Scatter Plot")
plt.show()
```
